#### 3. Escribir el código necesario para determinar de la manera más precisa el numéro de manzanas presentes en las imágenes. Utilizar algoritmos de selección de color, binazarización filtros morfológicos y determinación de regiones, entre otros, según sea necesario.

```python
import numpy as np
import cv2 as cv
import matplotlib. pyplot as plt
```

```python
img = cv.imread('manzanas1.jpg',1)
img2 = cv.imread('manzanas2.jpg',1)

img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)

plt.figure(figsize=(12,12))
plt.subplot(221)
plt.imshow(img, cmap='gray'), plt.xticks([]), plt.yticks([]);
plt.subplot(222)
plt.imshow(img2, cmap='gray'),plt.xticks([]), plt.yticks([]);
```

```python
#Se converte la imagen a HSV
img_hsv = cv.cvtColor(img, cv.COLOR_RGB2HSV)

# Definir el rango de colores rojos en HSV
lower_red = np.array([0,50,50])
upper_red = np.array([9,255,255])

mask_1 = cv.inRange(img_hsv, lower_red, upper_red)

#se crea una segunda máscara
lower_red2 = np.array([170,50,50])
upper_red2 = np.array([245,255,255])

mask_2 = cv.inRange(img_hsv, lower_red2, upper_red2)

mask = cv.bitwise_or(mask_1, mask_2)

# aplicar mascara
img_red = cv.bitwise_and(img,img, mask= mask)


plt.figure(figsize=(12,5))
plt.subplot(121)
plt.imshow(img, cmap='gray', vmin = 0, vmax=255)
plt.xticks([]), plt.yticks([]);

plt.subplot(122)
plt.imshow(img_red, cmap='gray', vmin = 0, vmax=255)
```

```python
# Aplicar filtrado morfológico para eliminar ruido

kernel = np.ones((10,10),np.uint8)# Kernel de 5x5

# Apertura para eliminar pequeños objetos blancos en el fondo rojo
apertura = cv.morphologyEx (mask, cv.MORPH_OPEN, kernel)

# Cierre para eliminar pequeños agujeros negros dentro de los objetos rojos
cierre = cv.morphologyEx(apertura, cv.MORPH_CLOSE, kernel)

plt.figure(figsize=(12,12))
plt.subplot(221)
plt.imshow(cierre, cmap='gray')
```

```python
# Se crea el Kernel o elementos estructurante
kernel = np.ones((3,3),np.uint8)
# se aplica el operador
erosion = cv.erode(cierre,kernel,iterations = 1)

plt.figure(figsize=(12,12))
plt.subplot(221)
plt.imshow(erosion, cmap='gray')
```

```python
# Obtener el número de etiquetas y la imagen de etiquetas
nEtiquetas, imgEtiquetas = cv.connectedComponents(erosion)

# Obtener el número máximo de etiqueta (excluyendo el fondo)
nEtiquetas = imgEtiquetas.max()
nEtiquetas
```

```python
plt.title('Regiones etiquetadas')
plt.xticks([])
plt.yticks([])
plt.imshow(imgEtiquetas)
print('Número de manzanas identificadas:', nEtiquetas)
```

```python
img_hsv2 = cv.cvtColor(img2, cv.COLOR_RGB2HSV)

# Definir el rango de colores rojos en HSV
lower_red = np.array([0,60,60])
upper_red = np.array([10,255,255])

mask2_1 = cv.inRange(img_hsv2, lower_red, upper_red)
#Se crea una segunda máscara
lower_red2 = np.array([160,60,60])
upper_red2 = np.array([200,255,255])

mask2_2 = cv.inRange(img_hsv2, lower_red2, upper_red2)

mask22 = cv.bitwise_or(mask2_1, mask2_2)


# aplicar mascara
img_red2 = cv.bitwise_and(img2,img2, mask= mask22)


plt.figure(figsize=(12,5))
plt.subplot(121)
plt.imshow(img2, cmap='gray', vmin = 0, vmax=255)
plt.xticks([]), plt.yticks([]);

plt.subplot(122)
plt.imshow(img_red2, cmap='gray', vmin = 0, vmax=255)
```
