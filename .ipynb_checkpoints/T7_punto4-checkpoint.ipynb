### 4. Aplicar los algoritmos necesarios para obtener en una imagen binaria, los detalles de las im√°genes de monedas adjunta.

```python
import numpy as np
import cv2 as cv
import matplotlib. pyplot as plt
```

```python
img = cv.imread('monedas.jpg',0)

plt.figure(figsize=(12,5))
plt.imshow(img, cmap='gray', vmin = 0, vmax=255)
plt.xticks([]), plt.yticks([]);
```

```python
# Aplicar umbralizaci√≥n para obtener una imagen binaria
_, binary_image = cv.threshold(img, 0, 255, cv.THRESH_BINARY_INV)
```
