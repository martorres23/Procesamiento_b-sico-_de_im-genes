```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('herramientas.jpg',0)
img2= cv.imread('utiles.jpg',0)
```

```python
# Al desplegar mediante pyplot, obtiene una ventana con una caja de herramientas.
plt.xticks([]), plt.yticks([])  # Oculta los valores en X e Y.
plt.imshow(img2, cmap = 'gray')
```

### Operador de intervalo de umbral binario


```python
#Se definen los umbrales
p1 = 100
p2 = 200

img_ubin = np.zeros_like(img, dtype=np.uint8)  # Crea una matriz de ceros del mismo tamaño que la imagen

img_ubin[(img <= p1) | (img >= p2)] = 255  # Asigna el valor 255 a los píxeles que cumplen el umbral
img_ubin[(img > p1) & (img < p2)] = 0  # Asigna el valor 0 a los píxeles que están entre los umbrales

img_ubin2 = np.zeros_like(img2, dtype=np.uint8)  # Crea una matriz de ceros del mismo tamaño que la imagen

img_ubin2[(img2 <= p1) | (img2 >= p2)] = 255  # Asigna el valor 255 a los píxeles que cumplen el umbral
img_ubin2[(img2 > p1) & (img2 < p2)] = 0  # Asigna el valor 0 a los píxeles que están entre los umbrales

plt.figure(figsize=(12,12))
plt.subplot(221)
plt.imshow(img_ubin, cmap='gray'), plt.xticks([]), plt.yticks([]);
plt.subplot(222)
plt.imshow(img_ubin2, cmap='gray'),plt.xticks([]), plt.yticks([]);
```

### Operador de intervalo de umbral binario invertido


```python
#Se definen los umbrales
p1 = 100
p2 = 200

img_ubini = np.zeros_like(img, dtype=np.uint8)  # Crea una matriz de ceros del mismo tamaño que la imagen

img_ubini[(img <= p1) | (img >= p2)] = 0  # Asigna el valor 0 a los píxeles que cumplen el umbral
img_ubini[(img > p1) & (img < p2)] = 255  # Asigna el valor 255 a los píxeles que están entre los umbrales

img_ubini2 = np.zeros_like(img2, dtype=np.uint8)  # Crea una matriz de ceros del mismo tamaño que la imagen

img_ubini2[(img2 <= p1) | (img2 >= p2)] = 0  # Asigna el valor 0 a los píxeles que cumplen el umbral
img_ubini2[(img2 > p1) & (img2 < p2)] = 255  # Asigna el valor 255 a los píxeles que están entre los umbrales

plt.figure(figsize=(12,12))
plt.subplot(221)
plt.imshow(img_ubini, cmap='gray'), plt.xticks([]), plt.yticks([]);
plt.subplot(222)
plt.imshow(img_ubini2, cmap='gray'),plt.xticks([]), plt.yticks([]);
```

### Operador de umbral de la escala de grises

```python
#Se definen los umbrales
p1 = 95
p2 = 215

img_ueg = np.zeros_like(img, dtype=np.uint8)  # Crea una matriz de ceros del mismo tamaño que la imagen

img_ueg[(img <= p1) | (img >= p2)] = 255  # Asigna el valor 255 a los píxeles que cumplen el umbral
img_ueg[(img > p1) & (img < p2)] = img[(img > p1) & (img < p2)]  # Asigna el valor p a los píxeles que están entre los umbrales

img_ueg2 = np.zeros_like(img2, dtype=np.uint8)  # Crea una matriz de ceros del mismo tamaño que la imagen

img_ueg2[(img2 <= p1) | (img2 >= p2)] = 255  # Asigna el valor 255 a los píxeles que cumplen el umbral
img_ueg2[(img2 > p1) & (img2 < p2)] = img2[(img2 > p1) & (img2 < p2)]  # Asigna el valor p a los píxeles que están entre los umbrales

plt.figure(figsize=(12,12))
plt.subplot(221)
plt.imshow(img_ueg, cmap='gray'), plt.xticks([]), plt.yticks([]);
plt.subplot(222)
plt.imshow(img_ueg2, cmap='gray'),plt.xticks([]), plt.yticks([]);
```

### Operador de umbral de la escala de grises invertido

```python
#Se definen los umbrales
p1 = 95
p2 = 215

img_uegi = np.zeros_like(img, dtype=np.uint8)  # Crea una matriz de ceros del mismo tamaño que la imagen

img_uegi[(img <= p1) | (img >= p2)] = 255  # Asigna el valor 255 a los píxeles que cumplen el umbral
img_uegi[(img > p1) & (img < p2)] = 255 - img[(img > p1) & (img < p2)]  # Asigna el valor 255-p a los píxeles que están entre los umbrales

img_uegi2 = np.zeros_like(img2, dtype=np.uint8)  # Crea una matriz de ceros del mismo tamaño que la imagen

img_uegi2[(img2 <= p1) | (img2 >= p2)] = 255  # Asigna el valor 255 a los píxeles que cumplen el umbral
img_uegi2[(img2 > p1) & (img2 < p2)] = 255 - img2[(img2 > p1) & (img2 < p2)]  # Asigna el valor 255-p a los píxeles que están entre los umbrales

plt.figure(figsize=(12,12))
plt.subplot(221)
plt.imshow(img_uegi, cmap='gray'), plt.xticks([]), plt.yticks([]);
plt.subplot(222)
plt.imshow(img_uegi2, cmap='gray'),plt.xticks([]), plt.yticks([]);
```

### Operador de extensión

```python
#Se definen los umbrales
p1 = 100
p2 = 200

img_oe = np.zeros_like(img, dtype=np.uint8)  # Crea una matriz de ceros del mismo tamaño que la imagen

img_oe[(img <= p1) | (img >= p2)] = 0  # Asigna el valor 0 a los píxeles que cumplen el umbral

# Asigna el valor (p - p1 * 255/p2 -p1) a los píxeles que están entre los umbrales
img_oe[(img > p1) & (img < p2)] = (img[(img > p1) & (img < p2)]-p1 *(255/p2)-p1)  

img_oe2 = np.zeros_like(img2, dtype=np.uint8)  # Crea una matriz de ceros del mismo tamaño que la imagen

img_oe2[(img2 <= p1) | (img2 >= p2)] = 0  # Asigna el valor 0 a los píxeles que cumplen el umbral

# Asigna el valor (p - p1 * 255/p2 -p1) a los píxeles que están entre los umbrales
img_oe2[(img2 > p1) & (img2 < p2)] = (img2[(img2 > p1) & (img2 < p2)]-p1 *(255/p2)-p1)  

plt.figure(figsize=(12,12))
plt.subplot(221)
plt.imshow(img_oe, cmap='gray'), plt.xticks([]), plt.yticks([]);
plt.subplot(222)
plt.imshow(img_oe2, cmap='gray'),plt.xticks([]), plt.yticks([]);
```

```python
# Aplicar la binarización utilizando un umbral de 128
_, img_bin = cv.threshold(img, 130, 255, cv.THRESH_BINARY)
_, img2_bin = cv.threshold(img2, 130, 255, cv.THRESH_BINARY)

# Se cambia el tamaño de las imagenes para que queden del mismo tamaño
imgg = cv.resize(img_bin, (1400, 1100))
imgg2 = cv.resize(img2_bin, (1400, 1100))
```

### Operadores lógicos: or, and y xor 

```python
# OPERADOR OR
img_or = np.logical_or(imgg, imgg2).astype(np.uint8) * 255

plt.xticks([]), plt.yticks([])
plt.imshow(img_or, cmap='gray')
plt.show()
```

```python
# OPERADOR AND
img_and = np.logical_and(imgg, imgg2).astype(np.uint8) * 255
plt.xticks([]), plt.yticks([])
plt.imshow(img_and, cmap='gray')
plt.show()
```

```python
# OPERADOR XOR
img_xor = np.logical_xor(imgg, imgg2).astype(np.uint8) * 255
plt.xticks([]), plt.yticks([])
plt.imshow(img_xor, cmap='gray')
plt.show()
```
